## Relationsships 

#### Person - student 
- A person may or may not be a student 
- A student must be exactly one person 

Person - instructor 
- A person may or may not be a instructor 
- An instructor must be exactly one person 

Person - ProgramManager 
- A person may or may not be a program manager 
- A program manager must be exactly one person 

#### Person - SensetivePersonData
- Every person has exactly one set of sensetive data 
- Sensetive data belongs to exactly one person 

#### Student - Enrollment 
- A student can have zero or many enrollments
- An enrollment must bbelong to exactly one student 

#### Class - Enrollment 
- A class can have zero or many enrollments 
- An enrollment must belog to exactly one class 

#### Program - Class
- A program has one or many classes (three intakes)
- A class belongs to exactly one program 

#### ProgramManager - Class 
- A program manager manages one or many classes 
- A class has exactly one program manager 

#### Program - Course 
A program consists of one or many courses 
- A course may belong to zero or many programs 

#### Program - ProgramCourse 
#### Course - ProgramCourse 

#### Instructor - Consultant 
- An instructor may or may not be a consultant 
- A consultant must be exactly one instructor 

#### Company - Consultant 
- A company can have zero or many consultants 
- A consultant belongs to exactly one company 

#### Campus - Class 
- A campus can host zero or many classes 
- A class belongs to exactly one campus
